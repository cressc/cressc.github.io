```python
import geocoder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import folium
```

```python
places=pd.read_excel("../ATstrategy/establishments.xlsx")
places
```

```python
m = folium.Map(location=[-29, 26],zoom_start=5)
               #'Stamen Terrain')
for i in range(0,len(places)):
#    print(i, places.lat[i],places.long[i])
    tooltip=places.name[i]   
    folium.CircleMarker(
    [places.lat[i],places.long[i]], popup=places.name[i], tooltip=tooltip, radius=3, color='red'
    ).add_to(m)

#places.lat[1]
m
```
